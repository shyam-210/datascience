
Excersice 3 :Bookscrap

```python
import csv
import requests
from bs4 import BeautifulSoup
url = 'https://books.toscrape.com/index.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
topic_links = soup.find('ul', class_='nav-list').find_all('a')
with open('book_data.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Topic', 'Book', 'Price', 'Stock Availability'])
    for link in topic_links:
        topic = link.text.strip()
        topic_url = url.replace('index.html', link['href'])
        topic_response = requests.get(topic_url)
        topic_soup = BeautifulSoup(topic_response.content, 'html.parser')
        book_containers = topic_soup.find_all('article', class_='product_pod')
        for book in book_containers:
            book_name = book.h3.a['title']
            book_price = book.find('p', class_='price_color').text.strip()
            stock_availability = book.find('p', class_='instock availability').text.strip()
            writer.writerow([topic, book_name, book_price, stock_availability])
print("Data has been saved to 'book_data.csv' file.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the climate data from a CSV file (assuming the data is in a CSV format)
data = pd.read_csv('climate_data.csv')

# Perform data classification and tabulation
# You can customize this based on your specific data columns and requirements
classification = data.groupby('Year')['Rainfall'].sum().reset_index()

# Print the tabulated data
print(classification)

# Plot the data graphically
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Rainfall', data=classification)
plt.title('Rainfall in Tamilnadu over the Years')
plt.xlabel('Year')
plt.ylabel('Rainfall (percentage)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset='sales data.csv'

datas=pd.read_csv(dataset)
#group datas based on total sales in a year
year=datas.groupby('Year')['Sales'].sum().reset_index()
#group datas based on average sales in a month in all 5 years
month=datas.groupby('Month')['Sales'].mean().reset_index()

print("The year wise total sales is")
print(year)
print("The month wise total sales is")
print(month)

year.plot(x="Year",y="Sales")
month.plot(x="Month",y="Sales")
```

Excercise 9 : CLASSIFICATION, CLUSTERING AND SVM TECHNIQUES FOR WEB SCRAPED DATASETS


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split

df = pd.read_csv("imdbdata.csv")
df.dropna(inplace=True)
le,svm = LabelEncoder(),SVR()
df['Certificate'] = le.fit_transform(df['Certificate'])
features,target = ['Certificate', 'Runtime', 'Meta_score', 'No_of_Votes'],'IMDB_Rating'
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
svm.fit(X_train, y_train)
certificate = input("Enter the certificate: ")
runtime = float(input("Enter the runtime (in minutes): "))
meta_score = float(input("Enter the meta score: "))
no_of_votes = int(input("Enter the number of votes: "))
data = pd.DataFrame([[certificate, runtime, meta_score, no_of_votes]], columns=['Certificate','Runtime','Meta_score','No_of_Votes'])
data['Certificate'] = le.transform(data['Certificate'])
imdb_rating_svm = svm.predict(data[features])
print("Predicted IMDb Rating:", imdb_rating_svm[0])
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

df=pd.read_csv('job data.csv')
inputs=df.drop(['salary more than 100k'],axis='columns')
target=df['salary more than 100k']

le=LabelEncoder()

inputs['company_n']= le.fit_transform(inputs['company'])
inputs['job_n']= le.fit_transform(inputs['job'])
inputs['degree_n']= le.fit_transform(inputs['degree'])
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
xtrain,xtest,ytrain,ytest=train_test_split(inputs_n,target,train_size=0.7)

model=DecisionTreeClassifier()

model.fit(xtrain,ytrain)
print("companies:\n\tFB is 0\tgoogle is 1\tWipro is 2")
print("Job:\n\tbusiness manager is 0\tprogrammer is 1\tsales executive is 2")
print("Degree:\n\tbachelor is 0\tmaster is 1")
salary=model.predict([[int(input("company:")),int(input("Job:")),int(input("degree:"))]])

print("salary is more than 100k") if(salary==1) else print("salary is less than 100k")
```
